## Load and Preprocess Data

### Subtask:

**Reasoning**:
First, I'll load the `Air_Quality.csv` dataset into a pandas DataFrame and display its first few rows and information to understand its structure and initial data types. This addresses the first instruction of loading the dataset and helps prepare for missing value inspection and treatment.



```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Air_Quality.csv')

# Display the first few rows of the DataFrame
print('First 5 rows of the dataset:')
print(df.head())

# Display information about the DataFrame to understand its structure and data types
print('\nDataFrame Info:')
df.info()
```

**Reasoning**:

## Identify Abnormal Pollution Patterns

### Subtask:
Apply an unsupervised learning technique, such as DBSCAN or Gaussian Mixture Models (GMM), to identify and flag abnormal pollution patterns within the preprocessed dataset. Visualize the detected anomalies.


**Reasoning**:
The previous step completed the data preprocessing and normalization. Now, I will apply the DBSCAN clustering algorithm to identify abnormal pollution patterns in the 'Normalized_Data_Value' feature, as per the instructions. This involves importing `DBSCAN`, fitting the model, and then creating a new column to store the anomaly labels. Subsequently, I'll prepare for visualization by importing `matplotlib.pyplot` and `seaborn`.



```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the 'Normalized_Data_Value' column and reshape it
X = df[['Normalized_Data_Value']]

# Instantiate DBSCAN model
# Using initial parameters: eps=0.05, min_samples=5. These might need tuning.
dbsc = DBSCAN(eps=0.05, min_samples=5)

# Fit the DBSCAN model to the data
df['Anomaly_DBSCAN'] = dbsc.fit_predict(X)

print('Distribution of Anomaly_DBSCAN labels:')
print(df['Anomaly_DBSCAN'].value_counts())

# Visualize the detected anomalies
plt.figure(figsize=(12, 7))
sns.scatterplot(
    x=df.index,
    y='Data Value',
    hue='Anomaly_DBSCAN',
    palette='viridis',
    data=df,
    s=100
)
plt.title('DBSCAN Anomaly Detection on Data Value')
plt.xlabel('Data Point Index')
plt.ylabel('Original Data Value')
plt.legend(title='DBSCAN Cluster (Anomaly = -1)')
plt.grid(True)
plt.show()
```

**Reasoning**:
The previous DBSCAN application with default parameters identified only two anomalies, which might not be representative of the actual abnormal pollution patterns. To get a more nuanced anomaly detection, I will now apply Gaussian Mixture Models (GMM) as an alternative unsupervised technique, as suggested by the subtask. GMM can model more complex data distributions and identify anomalies based on low probability density. I will visualize the results similar to DBSCAN to compare the outcomes.



```python
from sklearn.mixture import GaussianMixture
import numpy as np

# Initialize GMM with a suitable number of components (e.g., 2 for normal/abnormal)
# We'll use 2 components to capture different distributions in the data
gmm = GaussianMixture(n_components=2, random_state=42)

# Fit GMM to the 'Normalized_Data_Value'
gmm.fit(X)

# Predict the log-likelihood of each data point
# Data points with very low log-likelihood are considered anomalies
log_likelihood = gmm.score_samples(X)

# Define a threshold for anomaly detection (e.g., 1st percentile of log-likelihood)
# This threshold can be tuned. A lower percentile means stricter anomaly detection.
anomaly_threshold = np.percentile(log_likelihood, 1)

# Flag anomalies based on the threshold
df['Anomaly_GMM'] = (log_likelihood < anomaly_threshold).astype(int)

print('Distribution of Anomaly_GMM labels:')
print(df['Anomaly_GMM'].value_counts())

# Visualize the detected anomalies from GMM
plt.figure(figsize=(12, 7))
sns.scatterplot(
    x=df.index,
    y='Data Value',
    hue='Anomaly_GMM',
    palette='coolwarm',
    data=df,
    s=100
)
plt.title('GMM Anomaly Detection on Data Value')
plt.xlabel('Data Point Index')
plt.ylabel('Original Data Value')
plt.legend(title='GMM Anomaly (1 = Anomaly)')
plt.grid(True)
plt.show()
```

## Create Pollution Categories

### Subtask:
Based on a suitable air quality metric from the dataset, create a new categorical target variable with 'Low', 'Moderate', and 'High' pollution levels. This will be used for supervised learning.


**Reasoning**:
I will examine the descriptive statistics of the 'Data Value' column to understand its distribution and help determine appropriate thresholds for 'Low', 'Moderate', and 'High' pollution levels. This will inform the creation of the 'Pollution_Category' feature, which is the first step towards creating the target variable.



```python
print('Descriptive statistics for "Data Value":')
print(df['Data Value'].describe())
```

**Reasoning**:
Based on the descriptive statistics, I will now define thresholds using the 33rd and 66th percentiles of the 'Data Value' column. These thresholds will then be used to create the 'Pollution_Category' feature, categorizing pollution levels into 'Low', 'Moderate', and 'High', as required by the subtask.



```python
import numpy as np

# Calculate 33rd and 66th percentiles for 'Data Value'
p33 = df['Data Value'].quantile(0.33)
p66 = df['Data Value'].quantile(0.66)

print(f'33rd percentile (threshold for Low/Moderate): {p33:.2f}')
print(f'66th percentile (threshold for Moderate/High): {p66:.2f}')

# Define a function to categorize pollution levels
def categorize_pollution(value):
    if value <= p33:
        return 'Low'
    elif value <= p66:
        return 'Moderate'
    else:
        return 'High'

# Apply the categorization function to create the 'Pollution_Category' column
df['Pollution_Category'] = df['Data Value'].apply(categorize_pollution)

print('\nDistribution of Pollution_Category:')
print(df['Pollution_Category'].value_counts())

print('\nFirst 5 rows with new Pollution_Category:')
print(df.head())
```

## Evaluate Models and Visualize Results

### Subtask:
Evaluate the performance of each trained classifier using Accuracy, F1-score, and a confusion matrix. Visualize the confusion matrices and compare the metrics for all models, potentially including feature importance for tree-based models.


**Reasoning**:
I will import the necessary metrics from `sklearn.metrics` and plotting libraries (`matplotlib.pyplot`, `seaborn`). Then, for each model (Decision Tree, XGBoost, Neural Network), I will calculate and print the accuracy and weighted F1-score, and generate and visualize its confusion matrix using a heatmap. Additionally, I will extract and display the top 10 feature importances for the Decision Tree and XGBoost models, and finally, compile a summary of all models' performance metrics.



```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Get class labels from the LabelEncoder
class_labels = le.classes_

# Store metrics for comparison
model_performance = {}

# --- Decision Tree Model Evaluation ---
print('--- Decision Tree Classifier Evaluation ---')

# Accuracy
dt_accuracy = accuracy_score(y_test, y_pred_dt)
print(f'Accuracy: {dt_accuracy:.4f}')

# F1-score (weighted to account for potential class imbalance)
dt_f1 = f1_score(y_test, y_pred_dt, average='weighted')
print(f'F1-score (weighted): {dt_f1:.4f}')

model_performance['Decision Tree'] = {'Accuracy': dt_accuracy, 'F1-score': dt_f1}

# Confusion Matrix
cm_dt = confusion_matrix(y_test, y_pred_dt)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Decision Tree Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Feature Importance for Decision Tree
print('\n--- Decision Tree Feature Importance ---')
feature_importances_dt = pd.Series(dt_model.feature_importances_, index=X_train.columns)
top_10_features_dt = feature_importances_dt.nlargest(10)
print('Top 10 Most Important Features (Decision Tree):')
print(top_10_features_dt)

# --- XGBoost Model Evaluation ---
print('\n--- XGBoost Classifier Evaluation ---')

# Accuracy
xgb_accuracy = accuracy_score(y_test, y_pred_xgb)
print(f'Accuracy: {xgb_accuracy:.4f}')

# F1-score (weighted)
xgb_f1 = f1_score(y_test, y_pred_xgb, average='weighted')
print(f'F1-score (weighted): {xgb_f1:.4f}')

model_performance['XGBoost'] = {'Accuracy': xgb_accuracy, 'F1-score': xgb_f1}

# Confusion Matrix
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('XGBoost Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Feature Importance for XGBoost
print('\n--- XGBoost Feature Importance ---')
feature_importances_xgb = pd.Series(xgb_model.feature_importances_, index=X_train.columns)
top_10_features_xgb = feature_importances_xgb.nlargest(10)
print('Top 10 Most Important Features (XGBoost):')
print(top_10_features_xgb)

# --- Neural Network (MLP) Model Evaluation ---
print('\n--- Neural Network (MLP) Classifier Evaluation ---')

# Accuracy
mlp_accuracy = accuracy_score(y_test, y_pred_mlp)
print(f'Accuracy: {mlp_accuracy:.4f}')

# F1-score (weighted)
mlp_f1 = f1_score(y_test, y_pred_mlp, average='weighted')
print(f'F1-score (weighted): {mlp_f1:.4f}')

model_performance['Neural Network'] = {'Accuracy': mlp_accuracy, 'F1-score': mlp_f1}

# Confusion Matrix
cm_mlp = confusion_matrix(y_test, y_pred_mlp)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_mlp, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Neural Network (MLP) Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# --- Model Performance Summary ---
print('\n--- Model Performance Summary ---')
performance_df = pd.DataFrame(model_performance).T
print(performance_df)
```
